```python
teams = pd.read_csv("teams.csv") #we create an object called teams that reads in the csv file we uploaded by using pandas library to read
teams #call in the object to view the data that we uploaded
```

```python
teams = teams[["team", "country", "year", "athletes", "age", "prev_medals", "medals"]] #we decide to include only the unnecessary columns
teams #call the object again to view the dataset with only the necessary columns
```

```python
teams.corr()["medals"] #corr is a function in the pandas library that we use to find the CORRELATION of a specific column with the others
```

```python
import seaborn as sns #seaborn is a python data visulisation library built on top of matplotlib
#seaborn is widely used for data analysis and in this case we are using it for exploratory data visualisation tasks
```

```python
sns.lmplot(x="athletes", y="medals", data=teams, fit_reg=True, ci=None) #seaborn lmplot function is used to create a linear regression plot
#which provides a visual representation of the relationship between 2 variables and FITS a linear regression model to the data
```

```python
sns.lmplot(x="age", y="medals", data=teams, fit_reg=True, ci=None) #the ci parameter is used to control the confidence interval of the statistical
#estimation plotted. ci = none removes the confidence interval from the plot, resulting in a simpler visual representation
```

```python
teams.plot.hist(y="medals") #here i decided to plot a histogram and managed to infer that most countries will only have < 50 medals
```

```python
teams[teams.isnull().any(axis=1)] #this function is to find any existence of NULL values in each cell of teams
```

```python
teams = teams.dropna() #after confirming the existence of null values, i call dropna() to remove those said null values
teams
```

```python
train = teams[teams["year"] < 2012].copy() #creating the training set
test = teams[teams["year"] >= 2012].copy() #creating the test set
```

```python
train.shape #looking at the ration for training set
```

```python
test.shape #and test set, we can see that it's a nice 80-20 split conducive for our purposes
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression() #creating the linear regression class from the library sklearn
```

```python
predictors = ["athletes", "prev_medals"] #here we decide to use the athletes and prev_medals column as our predictors
target = "medals" #in order to predict for our target column which is to find out how many medals a country will get
```

```python
reg.fit(train[predictors], train["medals"]) #fitting the linear regression model with our predictors from the training set
```

```python
predictions = reg.predict(test[predictors]) #thereafter using it to predict for the test set with the predictors
```

```python
test["predictions"] = predictions #the predicted values we've created above are then assigned to a new column called prediction
test #and now we can see the the column predictions and compare
```

```python
test.loc[test["predictions"] < 0, "predictions"] = 0 #we use the locator function to find all the values under predictions which is less than 0
#then we turn them to 0 since a negative value for predictions brings us no use
```

```python
test
```

```python
test["predictions"] = test["predictions"].round() #we round off the values for predictions to the nearest whole number
```

```python
test
```

```python
from sklearn.metrics import mean_absolute_error
#importing mean_absolute_error function from sklearn library to calculate the MAE between the values

error = mean_absolute_error(test["medals"], test["predictions"])
#we call the MAE function to the actual values of the target variable medals and predictions as arguments and calculates the absolute differenc
#between the each pair of actual and predicted values
```

```python
error #after assigning we can now see the MAE value which is 3.29..
```

```python
teams.describe()["medals"] #looking at the medals column
#we can see that the error is way below the STD of 33, which means that our model is viable
```

```python
test[test["team"] == "USA"] #to look at TEAM USA and see how far our predictions are from the actual results
```

```python
test[test["team"] == "IND"] #in comparison we look at INDIA and see that for our 2012 prediction is good 6 vs 7 however for 2016 prediction
#we are off by 6x which is they got 2 models but we predicted 12..
```

```python
errors = (test["medals"] - test["predictions"]).abs() #abs is to find absolute value
errors # we can see the difference between how many medals we predicted and how many medals were won
```

```python
error_by_team = errors.groupby(test["team"]).mean() #to group it by team
error_by_team #we can now see how many medals we are off by team
```

```python
medals_by_team = test["medals"].groupby(test["team"]).mean() #looking at how many medals each country earned on average and group it by team
```

```python
error_ratio = error_by_team / medals_by_team # we find the ratio between our errors and medals by team
```

```python
error_ratio #and here we can see the ratio between our errors and medals
```

```python
error_ratio[~pd.isnull(error_ratio)] #this removes the valures that are NaN
```

```python
import numpy as np
error_ratio = error_ratio[np.isfinite(error_ratio)] #cleaning up our infinite values
error_ratio
```

```python
error_ratio.plot.hist() #looking at the error ratio histogram and see that our predictions are mostly okay, most of the predictions
#are about 0.5 from the actual
```

```python
error_ratio.sort_values() #from this we can see that countries that tend to get alot of medals in the olympics, our prediction error is lower
#and that for countries that dont get alot of medals in the olympics in the past, our error tend to get higher
```
